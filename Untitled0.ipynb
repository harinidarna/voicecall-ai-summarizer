```python
from google import genai
from google.genai import types # We import 'types' to access configuration settings

# 1. Initialize the client (remember to paste your key!)
client = genai.Client(api_key="PASTE_YOUR_API_KEY_HERE")

# 2. The same voice call transcript
transcript = """
Customer: Hi, I'm calling because my internet has been down since yesterday morning. I've restarted the router three times and the light is still blinking red.
Agent: I apologize for the inconvenience. Let me run a quick diagnostic on your line. It looks like there's an outage in your area that our technicians are currently working on.
Customer: Oh, okay. Do you know when it will be fixed? I work from home and really need it.
Agent: The estimated resolution time is within the next two hours. I can set up an automated text message to notify you once it's resolved.
Customer: Yes, please do that. Thank you.
"""

# 3. Update the prompt to explicitly ask for a specific JSON structure
prompt = f"""
Analyze this voice call transcript.
Return a JSON object with exactly two keys:
- 'summary': A short string summarizing the issue.
- 'action_items': A list of strings containing tasks for the agent.

Transcript:
{transcript}
"""

# 4. Call the model and force the output to be JSON
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt,
    config=types.GenerateContentConfig(
        response_mime_type="application/json",
    ),
)

# 5. Print the clean JSON payload
print(response.text)
```
