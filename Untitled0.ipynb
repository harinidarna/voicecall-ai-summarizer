```python
from google import genai
from google.genai import types # We import 'types' to access configuration settings

# 1. Initialize the client (remember to paste your key!)
client = genai.Client(api_key="PASTE_YOUR_API_KEY_HERE")

# 2. The same voice call transcript
transcript = """
Customer: Hi, I'm calling because my internet has been down since yesterday morning. I've restarted the router three times and the light is still blinking red.
Agent: I apologize for the inconvenience. Let me run a quick diagnostic on your line. It looks like there's an outage in your area that our technicians are currently working on.
Customer: Oh, okay. Do you know when it will be fixed? I work from home and really need it.
Agent: The estimated resolution time is within the next two hours. I can set up an automated text message to notify you once it's resolved.
Customer: Yes, please do that. Thank you.
"""

# 3. Update the prompt to explicitly ask for a specific JSON structure
prompt = f"""
Analyze this voice call transcript.
Return a JSON object with exactly two keys:
- 'summary': A short string summarizing the issue.
- 'action_items': A list of strings containing tasks for the agent.

Transcript:
{transcript}
"""

# 4. Call the model and force the output to be JSON
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt,
    config=types.GenerateContentConfig(
        response_mime_type="application/json",
    ),
)

# 5. Print the clean JSON payload
print(response.text)
```

```python
from google import genai
from google.genai import types

# 1. Initialize client
client = genai.Client("PASTE_YOUR_API_KEY_HERE")

# 2. Define the tool
def trigger_voice_call_flow(call_record_id: str, severity: str, action_required: str) -> dict:
    """Triggers an automated downstream flow for a voice call record based on customer sentiment."""
    print(f"\n>>> [TOOL EXECUTING] Routing Call ID: {call_record_id} | Severity: {severity}")
    return {"status": "success", "message": f"Flow triggered for {call_record_id}"}

# 3. The transcript
transcript = """
Agent: Thank you for calling support, this is Mark. Can I get your Call ID?
Customer: Yes, it's VC-9942. Listen Mark, I am absolutely furious. I was promised a refund three weeks ago and nothing has happened. I want this escalated to a manager immediately!
Agent: I am so sorry to hear that. Let me look into that for you...
"""

# 4. UPDATED PROMPT: Giving the AI room to "think out loud"
prompt = f"""
You are an integration engine processing voice call records to trigger automated flows.

Step 1: Write a brief, one-sentence analysis of the caller's sentiment.
Step 2: If the caller is highly frustrated or requests a manager, you MUST use the `trigger_voice_call_flow` tool.

Transcript:
{transcript}
"""

print("Analyzing transcript...\n")
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt,
    config=types.GenerateContentConfig(
        tools=[trigger_voice_call_flow],
        temperature=0.1,
    ),
)

# 5. Review the automatic execution history and final response
if response.automatic_function_calling_history:
    print("\n[SYSTEM LOG] The SDK automatically handled the following tool calls behind the scenes:")

    # We can loop through the history to see exactly what parameters the SDK extracted and used
    for turn in response.automatic_function_calling_history:
        for part in turn.parts:
            if part.function_call:
                print(f"- Automatically Executed: {part.function_call.name}")
                print(f"- Parameters Used: {part.function_call.args}")

# Print the final natural language summary
print(f"\nFinal AI Summary: {response.text}")
```
